## Nasal Cavity Skeleton

Data source: [Adaptation to Cold Climate in the Nasal Cavity Skeleton. A Comparison of Archaeological Crania from Different Climatic Zones \[PhD Dissertation\]](https://www.academia.edu/7382609/Adaptation_to_Cold_Climate_in_the_Nasal_Cavity_Skeleton._A_Comparison_of_Archaeological_Crania_from_Different_Climatic_Zones_PhD_Dissertation_) 




important features:

SERIES - series (Eskimo, Gr. Norse, Nubia, Med. Dane);

CHRONOL - chronology;

SEX - sex (male, female);

ANTER_AREA - pyriform aperture area;

POST_AREA - sum of choanae areas;

NOSE_ANGLE;

G_OP - maximal length of the skull;

EU_EU - maximal breadth of the skull;

BA_B - height of the skull;

N_PR - upper face height;

ZY_ZY - bizygomatic breadth;

ZM_ZM - medial face breadth;

FT_FT - minimal forehead breadth;

OL_ALV - ;

ENM_ENM - breadth of the palate;

N_NS - nasal height;

APT_APT - breadth of pyriform aperture;

GLOWNY - index = EU_EU/G_OP;

WYS_DL - index = BA_B/G_OP;
